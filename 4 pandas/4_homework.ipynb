```python
new_frame = all_cars.loc[lambda df: df['TG'] == 0, 'MIP']
new_frame.head()
list_l=[]
for d in new_frame:
    if d >=95.8984375 and d<=96.4140625:
        list_l.append(d)

        new_frame2 = all_cars.loc[lambda df: df['TG'] ==1, 'MIP']
new_frame2.head()
for d in new_frame2:
    if d >=77.4921875 and d<=83.7734375:
        list_l.append(d)
print(list_l)        
print(len(list_l))
```
