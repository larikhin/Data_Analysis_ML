```python
a = 0
tot_dict = {}
for v in list_names:
    tot_dict[v]=0
tot_dict
```

```python
for list_n in list_list:
    for val_from_list in list_n:
        for some_name in list_names:
            if val_from_list == some_name:
                tot_dict[val_from_list]=tot_dict[val_from_list]+1
tot_dict
```
